```python
#This notebook is used to generate Figure 1 from Reichl et al. (2024)
# submitted to ESS

import matplotlib.pyplot as plt
import xarray as xr
```

```python
Model = xr.open_dataset('../Data/MOM6Buoy_Outputs.nc')
```

```python
Lon='110w'
C='kr'
plt.figure(figsize=(3,4))
for ii,LBL in enumerate(['OM4','OM4up 225']):
    CL=C[ii]
    Zi=Model['Zi_'+LBL+'_'+Lon].values
    dZ = Zi[1:]-Zi[:-1]
    for zz in range(len(dZ)):
        plt.plot([dZ[zz],dZ[zz]],[Zi[zz],Zi[zz+1]],'-',color=CL)
        if zz>0:
            plt.plot([dZ[zz-1],dZ[zz]],[Zi[zz],Zi[zz]],'-',color=CL)
    plt.ylim(500,0)
    plt.xlim(0,60)
plt.xlabel('DZ [m]')
plt.ylabel('Depth [m]')
plt.savefig('../Fig/Resolution_vs_Depth.png',dpi=200,bbox_inches = "tight")
```
